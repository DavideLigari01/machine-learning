1) Leggere il dataset **AirBase_IT_v8_statistics.csv** (preferibilmente senza estrarlo dal file zip) che contiene le statistiche annuali per vari inquinanti

```python
import pandas as pd
dati=pd.read_csv(r'C:\Users\ligar\Desktop\esame algoritmi\AirBase_IT_v8_statistics.zip', sep='\t')
dati
```

```python
tmpDF=dati[(dati.statistic_shortname=='Max')]
tmpDF=tmpDF[['station_european_code','component_name','statistic_value','statistics_year']].copy()
maxValues=tmpDF.groupby(['station_european_code','component_name'])['statistic_value'].agg('max').reset_index()
result= maxValues.merge(tmpDF,on=['station_european_code','component_name','statistic_value'])
result.set_index(['station_european_code','component_name'], inplace=True)
result
```

3) Per ogni stazione, scoprire qual è l'inquinante il cui valore massimo ha avuto la maggiore variazione relativa nel corso degli anni ( 2 * (Massimo di Max - Minimo di Max) / (Massimo di Max + Minimo di Max)

```python
import numpy as np
minMax=tmpDF.groupby(['station_european_code','component_name'])['statistic_value'].agg([np.max,np.min]).reset_index()
minMax['varianza relativa']=2*((minMax['amax']-minMax['amin'])/(minMax['amax']+minMax['amin']))
tmp=minMax.groupby('station_european_code')['varianza relativa'].agg(np.max).reset_index()
result=tmp.merge(minMax,on=['station_european_code','varianza relativa'])
result=result[['station_european_code','component_name','varianza relativa']]
result.set_index(['station_european_code'], inplace=True)
result
```
